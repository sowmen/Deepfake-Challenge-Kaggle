```python
#print(logLoss(table))
def logLoss(data):
    loss = 0.0
    actual = data['actual']
    predicted =  data['predicted']
    total = len(actual)
    for i in range(total):
        loss += ((1.0  * actual[i]) * np.log(predicted[i]) + (1.0 * (1 - actual[i])) * np.log(1 - predicted[i]))
    print(loss)
    loss = (-1.0 * loss) / (1.0 * total)
    print("Loss: ")
    print(loss)
    return loss
    
```
